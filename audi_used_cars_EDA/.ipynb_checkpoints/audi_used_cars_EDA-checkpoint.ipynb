###### The following EDA is performed on Audi car dataset from [this link of kaggle]("https://www.kaggle.com/datasets/aishwaryamuthukumar/cars-dataset-audi-bmw-ford-hyundai-skoda-vw").

```python
import pandas as pd
data=pd.read_csv('audi.csv')
```

## EDA with the help of groupby method

```python
audi_model=data.groupby('model')['price'].agg(['count','mean','sum'])
audi_tran=data.groupby('transmission')['price'].agg(['count','mean','sum'])
audi_fuel=data.groupby('fuelType')['price'].agg(['count','mean','sum'])

# doing groupby operation as per required for the assignment
```

```python
audi_model=audi_model.sort_values(by='count',ascending=False)
audi_tran=audi_tran.sort_values(by='count',ascending=False)
audi_fuel=audi_fuel.sort_values(by='count',ascending=False)

# sorting all three dataframe by their share of count
```

```python
print("Top 5 selling car models:",(audi_model.index[:5]).tolist())
print("Top 5 selling transmission types:",(audi_tran.index[:5]).tolist())
print("Top 5 selling fueltypes cars:",(audi_fuel.index[:5]).tolist())
```

```python
print("Average selling price of top 5 selling cars based on models:",audi_model['mean'][:5].tolist())
print("Average selling price of top 5 selling cars based on Transmission types:",audi_tran['mean'][:5].tolist())
print("Average selling price of top 5 selling cars based on fueltype:",audi_fuel['mean'][:5].tolist())
```

```python
print("Total sale price of top 5 selling cars based on models:", audi_model['sum'][:5].tolist())
print("Total sale price of top 5 selling cars based on Transmission types:",audi_tran['sum'][:5].tolist())
print("Total sale price of top 5 selling cars based on Fueltypes:",audi_fuel['sum'][:5].tolist())
```

## EDA using Barplots

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(60,95))


plt.subplot(3,3,1)
sns.barplot(x='count',y=audi_model.index,data=audi_model,orient='h')
plt.title('\nTotal cars sold for each model of Audi\n',fontsize=55)
plt.xlabel('\nNumber of cars for each model\n\n',fontsize=40)
plt.ylabel('\nModels\n',fontsize=40)
plt.yticks(fontsize=32)
plt.xticks(fontsize=30)


plt.subplot(3,3,2)
sns.barplot(x='count',y=audi_tran.index,data=audi_tran,orient='h')
plt.title('\nTotal cars sold for each transmission type of Audi\n',fontsize=55)
plt.xlabel('\nNumber of cars for each transmission type\n\n',fontsize=40)
plt.ylabel('\nTransmissions\n',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)


plt.subplot(3,3,3)
sns.barplot(x='count',y=audi_fuel.index,data=audi_fuel,orient='h')
plt.title('\nTotal cars sold for each fuel type of Audi\n',fontsize=55)
plt.xlabel('\nNumber of cars for each fuel type\n\n',fontsize=40)
plt.ylabel('\nFuel Type\n',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)









plt.subplot(3,3,4)
sns.barplot(x='mean',y=audi_model.index,data=audi_model,orient='h')
plt.title('\nAverage selling price of Audi in terms of models cars\n',fontsize=55)
plt.xlabel('\nAverage selling price for each model\n\n',fontsize=40)
plt.ylabel('\nModels\n',fontsize=40)
plt.yticks(fontsize=32)
plt.xticks(fontsize=25)


plt.subplot(3,3,5)
sns.barplot(x='mean',y=audi_tran.index,data=audi_tran,orient='h')
plt.title('\nAverage selling price of Audi in terms of transmission type\n',fontsize=55)
plt.xlabel('\nAverage selling price for each transmission\n\n',fontsize=40)
plt.ylabel('\nTransmissions\n',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)


plt.subplot(3,3,6)
sns.barplot(x='mean',y=audi_fuel.index,data=audi_fuel,orient='h')
plt.title('\nAverage selling price of Audi in terms of Fueltype\n',fontsize=55)
plt.xlabel('\nAverage selling price for each Fueltype\n\n',fontsize=40)
plt.ylabel('\nFuel Type\n',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)










plt.subplot(3,3,7)
sns.barplot(x='sum',y=audi_model.index,data=audi_model,orient='h')
plt.title('Total sale of Audi in terms of models cars\n',fontsize=55)
plt.xlabel('\nTotal sale for each model',fontsize=40)
plt.ylabel('\nModels\n',fontsize=40)
plt.yticks(fontsize=32)
plt.xticks(fontsize=25)


plt.subplot(3,3,8)
sns.barplot(x='sum',y=audi_tran.index,data=audi_tran,orient='h')
plt.title('Total sale of Audi in terms of transmission type\n',fontsize=55)
plt.xlabel('\nTotal sale for each transmission',fontsize=40)
plt.ylabel('\nTransmission\n',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)


plt.subplot(3,3,9)
sns.barplot(x='sum',y=audi_fuel.index,data=audi_fuel,orient='h')
plt.title('Total sale of Audi in terms of Fueltype\n',fontsize=55)
plt.xlabel('\nTotal sale for each Fueltype',fontsize=40)
plt.ylabel('\nFuel Type\n',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)





plt.tight_layout()
plt.show()
```
