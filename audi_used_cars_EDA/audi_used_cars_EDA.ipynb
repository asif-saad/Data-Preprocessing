```python
import pandas as pd
```

```python
data=pd.read_csv("audi.csv")
```

```python
data.info()
```

```python
audi_model=data.groupby('model')['price'].agg(['count','mean','sum'])
```

```python
audi_tran=data.groupby('transmission')['price'].agg(['count','mean','sum'])
```

```python
audi_fuel=data.groupby('fuelType')['price'].agg(['count','mean','sum'])
```

```python
audi_model=audi_model.sort_values(by='count',ascending=False)
```

```python
audi_tran=audi_tran.sort_values(by='count',ascending=False)
```

```python
audi_fuel=audi_fuel.sort_values(by='count',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("darkgrid")
#plt.rcParams.update({'font.size': 35})
```

```python
plt.figure(figsize=(60,90))


plt.subplot(3,3,1)
sns.barplot(x='count',y=audi_model.index,data=audi_model,orient='h')
plt.title('Total cars sold for each model of Audi\n',fontsize=55)
plt.xlabel('\nNumber of cars for each model\n\n',fontsize=40)
plt.ylabel('Models',fontsize=40)
plt.yticks(fontsize=30)
plt.xticks(fontsize=30)



plt.subplot(3,3,2)
sns.barplot(x='count',y=audi_tran.index,data=audi_tran,orient='h')
plt.title('Total cars sold for each transmission type of Audi\n',fontsize=55)
plt.xlabel('\nNumber of cars for each transmission type\n\n',fontsize=40)
plt.ylabel('Transmissions',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)



plt.subplot(3,3,3)
sns.barplot(x='count',y=audi_fuel.index,data=audi_fuel,orient='h')
plt.title('Total cars sold for each fuel type of Audi\n',fontsize=55)
plt.xlabel('\nNumber of cars for each fuel type\n\n',fontsize=40)
plt.ylabel('Fuel Type',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)









plt.subplot(3,3,4)
sns.barplot(x='mean',y=audi_model.index,data=audi_model,orient='h')
plt.title('Average selling price of Audi in terms of models cars\n',fontsize=55)
plt.xlabel('\nAverage selling price for each model\n\n',fontsize=40)
plt.ylabel('Models',fontsize=40)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)


plt.subplot(3,3,5)
sns.barplot(x='mean',y=audi_tran.index,data=audi_tran,orient='h')
plt.title('Average selling price of Audi in terms of transmission type\n',fontsize=55)
plt.xlabel('\nAverage selling price for each transmission\n\n',fontsize=40)
plt.ylabel('Transmissions',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)


plt.subplot(3,3,6)
sns.barplot(x='mean',y=audi_fuel.index,data=audi_fuel,orient='h')
plt.title('Average selling price of Audi in terms of Fueltype\n',fontsize=55)
plt.xlabel('\nAverage selling price for each Fueltype\n\n',fontsize=40)
plt.ylabel('Fuel Type',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)










plt.subplot(3,3,7)
sns.barplot(x='sum',y=audi_model.index,data=audi_model,orient='h')
plt.title('Total sale of Audi in terms of models cars\n',fontsize=55)
plt.xlabel('\nTotal sale for each model',fontsize=40)
plt.ylabel('Models',fontsize=40)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)


plt.subplot(3,3,8)
sns.barplot(x='sum',y=audi_tran.index,data=audi_tran,orient='h')
plt.title('Total sale of Audi in terms of transmission type\n',fontsize=55)
plt.xlabel('\nTotal sale for each transmission',fontsize=40)
plt.ylabel('Transmissions',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)

plt.subplot(3,3,9)
sns.barplot(x='sum',y=audi_fuel.index,data=audi_fuel,orient='h')
plt.title('Total sale of Audi in terms of Fueltype\n',fontsize=55)
plt.xlabel('\nTotal sale for each Fueltype',fontsize=40)
plt.ylabel('Fuel Type',fontsize=40)
plt.yticks(rotation=90,fontsize=40)
plt.xticks(fontsize=30)





plt.tight_layout()
plt.show()
```
